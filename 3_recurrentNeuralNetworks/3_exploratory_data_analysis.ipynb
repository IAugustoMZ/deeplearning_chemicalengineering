# Brent Oil's Price Exploratory Analysis

```python
import warnings
import pandas as pd
from globals import *
import seaborn as sns
import matplotlib.pyplot as plt

# ignore warnings
warnings.filterwarnings('ignore')

# set ploting style
sns.set_style('darkgrid')
```

```python
# load data
data = pd.read_csv(DATASOURCE, parse_dates=['Date'])
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(1, 1, 1)
ax.plot(data['Date'], data['Price'])
ax.axvline(x=pd.to_datetime('01-10-2020'), lw=2, ls='--', color='red')
ax.set_ylabel('Brent Oil Price (USD / barrel)', size=24)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.set_title('Brent Oil Price Evolution - 1987 - 2022', size=28)
plt.savefig('../4_imgs/rnn_studies/eda_evoluation.png', dpi=500, bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(1, 1, 1)
ax.plot(data['Date'], data['Price'])
ax.axvline(x=pd.to_datetime('01-10-2020'), lw=2, ls='--', color='red')
ax.axvline(x=pd.to_datetime('01-01-2009'), lw=2, ls='--', color='orange')
ax.set_ylabel('Brent Oil Price (USD / barrel)', size=24)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.set_title('Brent Oil Price Evolution - 1987 - 2022', size=28)
plt.savefig('../4_imgs/rnn_studies/eda_evolution2.png', dpi=500, bbox_inches='tight')
```
